###3.5. Visualisasi Data Menggunakan Plot Univariat

```python
import matplotlib.pyplot as plt
df.plot(kind='box', figsize=(10,10), subplots=True, layout=(3,3), sharex=False, sharey=False)
# df.plot(kind='box')
plt.show()
```

Hasil dari perintah ini adalah sejumlah diagram box plot yang menampilkan distribusi dan statistik deskriptif dari setiap kolom numerik dalam DataFrame df. Setiap box plot menunjukkan kuartil pertama (Q1), median (Q2), dan kuartil ketiga (Q3).


Gambar tersebut menunjukkan serangkaian plot kotak yang menampilkan data kanker payudara Wisconsin. Plot kotak ini berdasarkan 10 fitur tumor payudara, yaitu:

-Clump thickness: Ketebalan gumpalan tumor

-Uniformity of cell size: Keseragaman ukuran sel

-Uniformity of cell shape: Keseragaman bentuk sel

-Margin adhesion: Adhesi tepi tumor

-Single epithelial cell size: Ukuran sel epitel tunggal

-Bland chromatin: Kromatin yang tidak bernoda

-Nucleoli: Nukleolus

-Mitoses: Mitosis

-Class: Kelas tumor (benign atau malignant)

Pada setiap plot kotak, nilai tengah (median) ditampilkan sebagai garis merah. Garis bawah plot kotak menunjukkan nilai kuartil pertama (Q1), dan garis atas plot kotak menunjukkan nilai kuartil ketiga (Q3). Garis penghubung antara Q1 dan Q3 menunjukkan rentang antarkuartil (IQR). Garis-garis vertikal di luar plot kotak menunjukkan nilai outlier.

Pada gambar tersebut, plot kotak untuk fitur clump thickness, uniformity of cell size, uniformity of cell shape, margin adhesion, single epithelial cell size, bland chromatin, dan mitoses memiliki nilai median yang lebih rendah untuk tumor malignant dibandingkan tumor benign. Hal ini menunjukkan bahwa fitur-fitur tersebut dapat digunakan untuk membedakan antara tumor benign dan malignant.

Plot kotak untuk fitur nucleoli memiliki nilai median yang lebih tinggi untuk tumor malignant dibandingkan tumor benign. Hal ini menunjukkan bahwa fitur nucleoli dapat digunakan untuk mendiagnosis tumor malignant.

Plot kotak untuk fitur class menunjukkan bahwa sebagian besar tumor (80%) adalah benign, sedangkan 20% adalah malignant.

Secara keseluruhan, gambar tersebut menunjukkan bahwa fitur-fitur tumor payudara dapat digunakan untuk membedakan antara tumor benign dan malignant. Fitur-fitur tersebut dapat digunakan untuk mendiagnosis tumor payudara dengan lebih akurat.
